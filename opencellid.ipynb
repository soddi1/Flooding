```python
import os
import pandas as pd
import matplotlib.pyplot as plt

file_path = r'E:\Desktop\LUMS\summer2024\Literature Review\data\410.csv\410.csv'

try:
    
    df = pd.read_csv(file_path)
    
    df['created_date'] = pd.to_datetime(df['created'], unit='s')
    df['updated_date'] = pd.to_datetime(df['updated'], unit='s')

    plt.figure(figsize=(12, 6))
    plt.hist(df['created_date'], bins=50, alpha=0.7, label='Created', color='blue')
    plt.hist(df['updated_date'], bins=50, alpha=0.7, label='Updated', color='green')
    plt.xlabel('Date')
    plt.ylabel('Number of Cell Towers')
    plt.title('Distribution of Cell Tower Installations and Updates Over Time')
    plt.legend()
    plt.grid(True)
    plt.show()

    output_file_path = r'E:\Desktop\LUMS\summer2024\Literature Review\data\410_converted.csv'  # Change this to the desired output path
    df.to_csv(output_file_path, index=False)

except PermissionError as e:
    print(f"PermissionError: {e}")
except FileNotFoundError as e:
    print(f"FileNotFoundError: {e}")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

file_path = r'E:\Desktop\LUMS\summer2024\Literature Review\data\410.csv\410.csv'

sindh_lat_min, sindh_lat_max = 23.6, 28.6  
sindh_lon_min, sindh_lon_max = 66.8, 71.2  

try:
    df = pd.read_csv(file_path)
    
    df['created_date'] = pd.to_datetime(df['created'], unit='s')
    df['updated_date'] = pd.to_datetime(df['updated'], unit='s')

    sindh_df = df[(df['lat'] >= sindh_lat_min) & (df['lat'] <= sindh_lat_max) &
                  (df['lon'] >= sindh_lon_min) & (df['lon'] <= sindh_lon_max)]

    plt.figure(figsize=(12, 6))
    plt.hist(sindh_df['created_date'], bins=50, alpha=0.7, label='Created', color='blue')
    plt.hist(sindh_df['updated_date'], bins=50, alpha=0.7, label='Updated', color='green')
    plt.xlabel('Date')
    plt.ylabel('Number of Cell Towers')
    plt.title('Distribution of Cell Tower Installations and Updates Over Time in Sindh')
    plt.legend()
    plt.grid(True)
    plt.show()
    output_file_path = r'E:\Desktop\LUMS\summer2024\Literature Review\data\410_sindh.csv'
    sindh_df.to_csv(output_file_path, index=False)

except PermissionError as e:
    print(f"PermissionError: {e}")
except FileNotFoundError as e:
    print(f"FileNotFoundError: {e}")
except Exception as e:
    print(f"An error occurred: {e}")
```
