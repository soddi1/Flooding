```python
import requests
import json
import time
import matplotlib.pyplot as plt
from datetime import datetime

# Parameters
asn = 9541
sampling_period = 3600  # 1 day in seconds

# Helper function to fetch data for a specific time range
def fetch_bgp_update_data(start_time, end_time):
    url = f"https://stat.ripe.net/data/bgp-update-activity/data.json?resource={asn}&starttime={start_time}&endtime={end_time}&sampling_period={sampling_period}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()['data']['updates']
    else:
        print(f"Failed to fetch data: {response.status_code}")
        return []

# Define start and end times for each month
time_ranges = [
    (1640995200, 1643673599),  # January 2022
    (1643673600, 1646092799),  # February 2022
    (1646092800, 1648771199),  # March 2022
    (1648771200, 1651363199),  # April 2022
    (1651363200, 1654041599),  # May 2022
    (1654041600, 1656633599),  # June 2022
    (1656633600, 1659311999),  # July 2022
    (1659312000, 1661990399),  # August 2022
    (1661990400, 1664582399),  # September 2022
    (1664582400, 1667174399),  # October 2022
    (1667174400, 1669852799),  # November 2022
    (1669852800, 1672444799),   # December 2022
    (1672444800, 1675123199),  # January 2023
    (1675123200, 1677542399)   # February 2023
]
# Fetching data for each time period
all_updates = []
for start_time, end_time in time_ranges:
    updates = fetch_bgp_update_data(start_time, end_time)
    all_updates.extend(updates)
    time.sleep(1)  # Be polite to the server

# Save the combined data to a file
output_file = 'bgp_update_data_cybernet.json'
with open(output_file, 'w') as f:
    json.dump(all_updates, f, indent=2)

print(f"Data saved to {output_file}")

with open(output_file, 'r') as f:
    bgp_update_data = json.load(f)

# Extract timestamps and number of announcements
timestamps = [datetime.fromisoformat(update['starttime']) for update in bgp_update_data]
num_announcements = [update['announcements'] for update in bgp_update_data]

# Plot the data
plt.figure(figsize=(10, 5))
plt.plot(timestamps, num_announcements, label='Number of Announcements')
plt.xlabel('Time')
plt.ylabel('Number of Announcements')
plt.title(f'BGP Announcements Over Time for ASN {asn}')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
